```python
def load_words(path: str) -> list:
    words = []
    with open(path, "r") as f:
        words = [w.upper().strip() for w in f.readlines()[:100]]
        print(len(words))
    return words
path = "/Users/anika/words.txt"
difficulty_scores = [8, 4, 9, 5, 8, 9, 2, 9, 3, 7, 8, 7, 2, 9, 4, 5, 3, 3, 9, 5, 3, 10, 6, 3, 4, 10, 6, 9, 8, 2, 9, 9, 7, 7, 8, 8, 3, 8, 10, 6, 7, 6, 7, 8, 9, 4, 10, 9, 7, 7, 9, 9, 5, 4, 9, 6, 8, 8, 5, 8, 7, 9, 1, 2, 9, 5, 8, 8, 2, 4, 10, 8, 5, 2, 8, 10, 9, 9, 10, 4, 3, 2, 10, 5, 7, 3, 7, 8, 5, 9, 8, 8, 8, 9, 10, 7, 5, 2, 9, 8]

words = load_words(path)
assert len(words) == len(difficulty_scores)
```

```python
def word_to_features(word): # convert a word into numerical features

    # feature 1: letter frequency score
    letter_freq_map = {
        'E': 12.7, 'T': 9.1, 'A': 8.2, 'O': 7.5, 'I': 7.0,
        'N': 6.7, 'S': 6.3, 'H': 6.1, 'R': 6.0, 'D': 4.3,
        'L': 4.0, 'C': 2.8, 'U': 2.8, 'M': 2.4, 'W': 2.4,
        'F': 2.2, 'G': 2.0, 'Y': 2.0, 'P': 1.9, 'B': 1.5,
        'V': 1.0, 'K': 0.8, 'J': 0.2, 'X': 0.2, 'Q': 0.1, 'Z': 0.1
    }
    avg_freq = n.mean([letter_freq_map[letter] for letter in word])

    # feature 2:
    vowels = len([v for v in word if v in 'AEIOU'])

    # feature 3: has_repeated
    has_repeat = len(set(word)) != len(word)

    # feature 4: num_unique
    num_unique = len(set(word))

    return np.array([avg_freq, vowels, has_repeat, num_unique])  # what do we need to predict..

# convert all words to features
# independent variable features, dependent variable difficulty
X = np.array([word_to_features(w) for w in words])  # remember list comprehension!!
y = np.array(difficulty_scores)

print(X.shape)
print(y.shape)
print(f"word: {words[6]}")
print(f"features: {X[6]}")
print(f"difficulty: {y[6]} ")
```

```python
def train_test_split(X, y, test_size=0.1):
    """Split data into train and test sets"""
    n = len(X)
    n_test = int(n * test_size)

    indices = np.random.permutation(n)
    test_indices = indices[:n_test]
    train_indices = indices[n_test:]

    X_train = X[train_indices]
    X_test = X[test_indices]
    y_train = y[train_indices]
    y_test = y[test_indices]
    
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test_split(X, y)

# ADD THESE DEBUG LINES:
print("\n=== SPLIT DEBUG INFO ===")
print(f"Original X shape: {X.shape}")
print(f"Original y shape: {y.shape}")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
print(f"X_train samples: {len(X_train)}")
print(f"X_test samples: {len(X_test)}")  # ‚Üê This is probably 0!

print(f"Training samples: {len(X_train)}")
print(f"Test samples: {len(X_test)}")
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.0001, n_iterations=200):
        self.lr = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        self.losses = []  # track loss over time

    def fit(self, X, y):
        """Train the model using gradient descent"""
        n_samples, n_features = X.shape

        # Initialize weights and biases
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient descent
        for i in range(self.n_iterations):
            # forward pass: predictions
            y_pred = X @ self.weights + self.bias
            
            # calculate loss (MSE)
            mse = np.mean((y_pred - y) ** 2)
            self.losses.append(mse)
            
            # calculate gradients
            dw = (2 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (2 / n_samples) * np.sum(y_pred - y)

            # update weights
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
            # print progress every 100 iterations
            if (i + 1) % 100 == 0:
                print(f"Iteration {i+1}/{self.n_iterations}, Loss: {mse:.4f}")
        print(self.weights)
        print(self.bias)

    def predict(self, X):
        """Make predictions"""
        return np.dot(X, self.weights) + self.bias  # is bias like offset

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# evaluate - how well would it do?
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_mse = np.mean((y_train_pred - y_train) ** 2)
test_mse = np.mean((y_test_pred - y_test) ** 2)

print(f"{train_mse=}, {test_mse=}")

print("\n=== DETAILED TEST ANALYSIS ===")
for i in range(len(X_test)):
    # Find which word this is
    for j, features in enumerate(X):
        if np.allclose(X[j], X_test[i]):
            word = words[j]
            break
    
    pred = y_test_pred[i]
    actual = y_test[i]
    
    print(f"Word: {word:10} | Features: {X_test[i]}")
    print(f"  Predicted: {pred:.2f} | Actual: {actual:.2f} | Error: {pred - actual:.2f}")
    print()

# Calculate MSE manually to see where NaN comes from
errors_squared = (y_test_pred - y_test) ** 2
print(f"Squared errors: {errors_squared}")
print(f"Any NaN in squared errors? {np.isnan(errors_squared).any()}")
```

```python
# confusion matrices
y_true = np.array([1, 0, 1, 1, 0, 0, 1])
y_pred = np.array([1, 0, 1, 0, 0, 1, 1])

tp = ((y_true==1) & (y_pred==1)).sum()
fp = ((y_true==0) & (y_pred==1)).sum()
fn = ((y_true==1) & (y_pred==1)).sum()
tn = ((y_true==0) & (y_true==0)).sum()
# note: & and | are bitwise ops but numpy overloads them to mean element logical ops on arrays

precision = tp/(tp+fp)  # of positives, how many acc real? (conversaervative)
recall = tp/(tp+fn)  # how many of acc positivs? bc false engativs meant y_true = 1 
f1 = 2 * precision * recall / (precision + recall)  # how is this good?
print(f"{precision=}, {recall=}, {f1=}")
```

```python
import torch
import torch.nn as nn

# fake data: [age, shoe_size] -> height
X = torch.tensor([[10.0, 6.0],
                  [15.0, 8.0],
                  [20.0, 9.0],
                  [25.0, 10.0]])
y = torch.tensor([[120.0], [150.0], [170.0], [180.0]])

# build the model
# model = nn.Linear(in_features=2, out_features=1)  # linear regression
# loss_fn = nn.MSELoss()
# optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

model = nn.Linear(in_features=2, out_features =1)
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)

# training loop
num_epochs = 500
for epoch in range(num_epochs):
    y_pred = model(X)
    loss = loss_fn(y_pred, y)
    optimizer.zero_grad()  # zero gradients
    loss.backward()  # backward, auto compute gradients
    optimizer.step()  # update weights
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: loss={loss.item():.4f}")


new_data = torch.tensor([18,9], dtype=torch.float32)
print(f"predicted height: {model(new_data).item()}")


    
```
